## Import the Required Libraries

```python
import numpy as np #Linear Algebra and maths
import matplotlib.pyplot as plt # General plotting library
import matplotlib as mpl
import ipywidgets as widgets #interactivity widgets
from ipywidgets import interact, interactive, fixed, interact_manual
import warnings # To supress warnings that may showup due to version incompatibility
warnings.filterwarnings('ignore')
%config InlineBackend.figure_format = 'svg'
```

## Create static sine and cosine waves

```python
x=np.arange(0,2*np.pi,0.1) #arange defines start, end and increment. 
y=np.sin(x)
plt.plot(x,y)
y2=np.cos(x)
plt.plot(x,y2)
```

## Lets plot the same graphs with Lables

```python
x=np.arange(0,2*np.pi,0.1) #arange defines start, end and increment. 
y=np.sin(x)
plt.plot(x,y)
y2=np.cos(x)
plt.plot(x,y2)
## The following is the lable block
plt.xlabel('x values from 0 to 2 pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Graph of Sine and Cosine from 0 to 2pi') # To display the title
plt.legend(['sin(x)','cos(x)'])
```

## Let's Create a Ricker wavelet (static)

```python
def ricker(f,length=0.512,dt=0.001):
    t=np.linspace(-length/2,(length-dt)/2,length/dt)
    y=(1.-2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    return t, y
f=5 # Lets start with a frequencey of 5 hz
t,y=ricker(f)
plt.plot(t,y)
```

## I will take it a step further and add an interactive slider to control the frequency and change the ricker plot on the fly

```python
##Lets set the length and dt upfront
length=0.512;
dt=0.001;
def ricker(f):
    t=np.linspace(-length/2,(length-dt)/2,length/dt)
    y=(1.-2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    plt.plot(t,y,c="green")
    plt.xlabel("value")
    plt.ylabel("Amplitude")
#Interactivity block
interact(ricker,f=widgets.FloatSlider(value=7.5,
                                     min=1,
                                     max=10.0,
                                     step=0.5))
```
